```python
#API PROJECT
#import librairies

import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#Define the initial API endpoints to work with

url='https://www.themealdb.com/api/json/v1/1/categories.php'
response = requests.get(url)
results = response.json()

#print the results
results
```

```python
#Normalize our results
flatten_data = json_normalize(results)
flatten_data
```

```python
flatten_data = pd.DataFrame(results)
flatten_data.head(10)
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data

```

```python
nested_data= ['categories']
flat= flatten(flatten_data, nested_data)
flat
```

```python
#Rename columns ID
flat= flat.rename(columns={"categories_idCategory": "ID_Category", "categories_strCategory": "Type of food", 'categories_strCategoryDescription': 'Description', 'categories_strCategoryThumb': 'Link'})
flat
```
